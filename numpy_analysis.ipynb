```python
import numpy as np

# 1D Array
arr1 = np.array([10, 20, 30, 40, 50])
print("1D Array:", arr1)

# 2D Array
arr2 = np.array([[1, 2], [3, 4]])
print("\n2D Array:")
print(arr2)

# Vectorized Operations
print("\nAdd 5:", arr1 + 5)
print("Multiply by 2:", arr1 * 2)

# Statistics
print("\nMean:", np.mean(arr1))
print("Standard Deviation:", np.std(arr1))
print("50th Percentile:", np.percentile(arr1, 50))

# Boolean Masking
print("\nValues Greater Than 25:")
print(arr1[arr1 > 25])
```
