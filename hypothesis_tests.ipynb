```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
print("Scenario 1")
print("H0: Average customer spending is ₹500.")
print("H1: Average customer spending is not ₹500.")

print("\nScenario 2")
print("H0: Sales of Store A and Store B are equal.")
print("H1: Sales of Store A and Store B are different.")

print("\nScenario 3")
print("H0: New marketing campaign has no effect.")
print("H1: New marketing campaign increases sales.")
```

```python
sales = [510,495,505,500,490,515,520,500,498,502]

t_stat, p_value = stats.ttest_1samp(sales, 500)

print("One Sample t-test")
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
store_a = [500,510,520,515,505]
store_b = [490,495,500,485,492]

t_stat, p_value = stats.ttest_ind(store_a, store_b)

print("Two Sample t-test")
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
confidence = stats.t.interval(
    0.95,
    len(sales)-1,
    loc=np.mean(sales),
    scale=stats.sem(sales)
)

print("95% Confidence Interval:")
print(confidence)
```
