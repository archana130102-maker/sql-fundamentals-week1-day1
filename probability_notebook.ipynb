```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Contingency Table

data = {
    "Male": [30, 20],
    "Female": [25, 25]
}

table = pd.DataFrame(data, index=["Purchased", "Not Purchased"])

print(table)

total = table.values.sum()

P_A = table.loc["Purchased"].sum() / total
P_B = table["Male"].sum() / total
P_A_and_B = table.loc["Purchased", "Male"] / total
P_A_given_B = P_A_and_B / P_B

print("P(A):", P_A)
print("P(B):", P_B)
print("P(A∩B):", P_A_and_B)
print("P(A|B):", P_A_given_B)
```

```python
x = np.linspace(-4, 4, 100)

y = norm.pdf(x)

plt.figure(figsize=(8,5))
plt.plot(x, y)
plt.title("Normal Distribution")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

```python
n = 10
p = 0.5

samples = np.random.binomial(n, p, 1000)

plt.hist(samples, bins=11)
plt.title("Binomial Distribution")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
means = []

for i in range(1000):
    sample = np.random.randint(1, 101, 30)
    means.append(np.mean(sample))

plt.hist(means, bins=30)
plt.title("Central Limit Theorem")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.show()
```
